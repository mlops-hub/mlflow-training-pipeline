## Explore Data - Data Analysis - EDA

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import os
```

```python
PROJECT_ROOT = os.path.dirname(os.path.dirname(os.getcwd()))

MERGED_DATASET_PATH = os.path.join(PROJECT_ROOT, "data/ingestion/merged_df.csv")

df = pd.read_csv(MERGED_DATASET_PATH)
```

```python
df.shape
```

```python
df.info()
```

```python
df.describe(include='all')
```

```python
df.isnull().sum().sort_values(ascending=False)
```

```python
df.duplicated().sum()
```

```python
for col in df.columns:
        print(f'{col}: ', df[col].nunique())
```

```python
sns.boxplot(data=df, x='legs')
plt.show()
```

```python
# Correlation Heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm', fmt='.2f')
plt.show()
```

```python
# save merged_df 
output_dir = os.path.join(PROJECT_ROOT ,"data/eda")
os.makedirs(output_dir, exist_ok=True)

df.to_csv(f"{output_dir}/eda_df.csv", index_label=False)
```
